### Task 1: Visualizing Data Distributions


##### 1. Histogram of `ConvertedCompYearly`


Visualize the distribution of yearly compensation (`ConvertedCompYearly`) using a histogram.



```python
df1 = df[['ConvertedCompYearly']].copy()
df1['ConvertedCompYearly'] = pd.to_numeric(df1['ConvertedCompYearly'], errors='coerce')
df1 = df1.dropna()
df1 = df1[df1['ConvertedCompYearly'] < 1000000]

plt.figure(figsize=(10,6))
plt.hist(df1['ConvertedCompYearly'], bins=30)
plt.title('Distribution of ConvertedCompYearly')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('Frequency')
plt.show()
```

##### 2. Box Plot of `Age`


Since `Age` is categorical in the dataset, convert it to numerical values for a box plot.



```python
age_counts = df['Age'].value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.bar(age_counts.index, age_counts.values)
plt.title('Distribution of Age Groups')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Task 2: Visualizing Relationships in Data


##### 1. Scatter Plot of `Age_numeric` and `ConvertedCompYearly`


Explore the relationship between age and compensation.



```python
df_scatter = df[['Age', 'ConvertedCompYearly']].copy()

df_scatter['Age_numeric'] = df_scatter['Age'].str.extract('(\d+)').astype(float)
df_scatter['ConvertedCompYearly'] = pd.to_numeric(df_scatter['ConvertedCompYearly'], errors='coerce')

df_scatter = df_scatter.dropna(subset=['Age_numeric', 'ConvertedCompYearly'])
df_scatter = df_scatter[df_scatter['ConvertedCompYearly'] < 1000000]

plt.figure(figsize=(10,6))
plt.scatter(df_scatter['Age_numeric'], df_scatter['ConvertedCompYearly'], alpha=0.5)
plt.title('Age vs Compensation')
plt.xlabel('Age')
plt.ylabel('ConvertedCompYearly')
plt.show()
```

##### 2. Bubble Plot of `ConvertedCompYearly` and `JobSatPoints_6` with `Age_numeric` as Bubble Size


Explore how compensation and job satisfaction are related, with age as the bubble size.


```python
df_bubble = df[['Age', 'ConvertedCompYearly', 'JobSatPoints_6']].copy()

df_bubble['Age_numeric'] = df_bubble['Age'].str.extract('(\d+)').astype(float)
df_bubble['ConvertedCompYearly'] = pd.to_numeric(df_bubble['ConvertedCompYearly'], errors='coerce')
df_bubble['JobSatPoints_6'] = pd.to_numeric(df_bubble['JobSatPoints_6'], errors='coerce')

df_bubble = df_bubble.dropna(subset=['Age_numeric', 'ConvertedCompYearly', 'JobSatPoints_6'])
df_bubble = df_bubble[df_bubble['ConvertedCompYearly'] < 1000000]

plt.figure(figsize=(10,6))
plt.scatter(
    df_bubble['ConvertedCompYearly'],
    df_bubble['JobSatPoints_6'],
    s=df_bubble['Age_numeric'] * 2,
    alpha=0.5
)

plt.title('Compensation vs Job Satisfaction (Bubble = Age)')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('JobSatPoints_6')
plt.show()
```

### Task 3: Visualizing Composition of Data with Bar Charts


##### 1. Horizontal Bar Chart of `MainBranch` Distribution


Visualize the distribution of respondents’ primary roles to understand their professional focus.



```python
mainbranch_counts = df['MainBranch'].value_counts()

plt.figure(figsize=(10,6))
plt.barh(mainbranch_counts.index, mainbranch_counts.values)
plt.title('MainBranch Distribution')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.show()
```

##### 2. Vertical Bar Chart of Top 5 Programming Languages Respondents Want to Work With


Identify the most desired programming languages based on `LanguageWantToWorkWith`.



```python
df_lang = df[['LanguageWantToWorkWith']].dropna().copy()

df_lang['Language'] = df_lang['LanguageWantToWorkWith'].str.split(';')
df_lang = df_lang.explode('Language')
df_lang['Language'] = df_lang['Language'].str.strip()

top_lang = df_lang['Language'].value_counts().head(5)

plt.figure(figsize=(10,6))
plt.bar(top_lang.index, top_lang.values)
plt.title('Top 5 Programming Languages Respondents Want To Work With')
plt.xlabel('Language')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

##### 3. Stacked Bar Chart of Median `JobSatPoints_6` and `JobSatPoints_7` by Age Group


Compare job satisfaction metrics across different age groups with a stacked bar chart.


```python
df_stack = df[['Age', 'JobSatPoints_6', 'JobSatPoints_7']].copy()

df_stack['JobSatPoints_6'] = pd.to_numeric(df_stack['JobSatPoints_6'], errors='coerce')
df_stack['JobSatPoints_7'] = pd.to_numeric(df_stack['JobSatPoints_7'], errors='coerce')

df_stack = df_stack.dropna()

grouped = df_stack.groupby('Age')[['JobSatPoints_6', 'JobSatPoints_7']].median()

grouped.plot(kind='bar', stacked=True, figsize=(12,6))

plt.title('Median Job Satisfaction by Age Group')
plt.xlabel('Age')
plt.ylabel('Median Score')
plt.xticks(rotation=45)
plt.show()
```

##### 4. Bar Chart of Database Popularity (`DatabaseHaveWorkedWith`)


Identify the most commonly used databases among respondents by visualizing `DatabaseHaveWorkedWith`.



```python
df_db = df[['DatabaseHaveWorkedWith']].dropna().copy()

df_db['Database'] = df_db['DatabaseHaveWorkedWith'].str.split(';')
df_db = df_db.explode('Database')
df_db['Database'] = df_db['Database'].str.strip()

top_db = df_db['Database'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(top_db.index, top_db.values)
plt.title('Database Popularity')
plt.xlabel('Database')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Task 4: Visualizing Comparison of Data with Bar Charts


##### 1. Grouped Bar Chart of Median `ConvertedCompYearly` for Different Age Groups


Compare median compensation across multiple age groups with a grouped bar chart.



```python
df_age = df[['Age', 'ConvertedCompYearly']].copy()

df_age['ConvertedCompYearly'] = pd.to_numeric(df_age['ConvertedCompYearly'], errors='coerce')
df_age = df_age.dropna()

median_comp = df_age.groupby('Age')['ConvertedCompYearly'].median().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(median_comp.index, median_comp.values)
plt.title('Median Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation')
plt.xticks(rotation=45)
plt.show()
```

##### 2. Bar Chart of Respondent Count by Country


Show the distribution of respondents by country to see which regions are most represented.



```python
df_country = df[['Country']].dropna().copy()

top_countries = df_country['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(top_countries.index, top_countries.values)
plt.title('Top 10 Countries by Respondent Count')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
